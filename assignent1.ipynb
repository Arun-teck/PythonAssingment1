```python
def strings(str1, str2):
    while str2:
        str1, str2 = str2, str1[:len(str1) % len(str2)]
    return str1
output = strings("ABCABC","ABC")
print("Output:",output)
```

```python
def strings(str1, str2):
    from math import gcd
    return str1[:gcd(len(str1), len(str2))] if str1 + str2 == str2 + str1 else ""
output = strings("LEET","CODE")
print("Output:",output)
```
